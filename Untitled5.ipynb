```python
from transformers import VisionEncoderDecoderModel, ViTFeatureExtractor, AutoTokenizer
from PIL import Image
import requests
import torch
from IPython.display import display  # Jupyter/Colab用

# モデル、トークナイザー、特徴抽出器の準備
model = VisionEncoderDecoderModel.from_pretrained("nlpconnect/vit-gpt2-image-captioning")
feature_extractor = ViTFeatureExtractor.from_pretrained("nlpconnect/vit-gpt2-image-captioning")
tokenizer = AutoTokenizer.from_pretrained("nlpconnect/vit-gpt2-image-captioning")

# 推論用デバイスの設定（GPUが利用可能な場合はGPUを使用）
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# キャプションを生成する関数
def generate_caption(image_url):
    # 画像の読み込み
    image = Image.open(requests.get(image_url, stream=True).raw)

    # 画像を表示する
    display(image)  # Jupyter/Colab用に画像を表示

    # 画像をモデルに適した形式に変換
    pixel_values = feature_extractor(images=image, return_tensors="pt").pixel_values
    pixel_values = pixel_values.to(device)

    # キャプションの生成
    output_ids = model.generate(pixel_values, max_length=16, num_beams=4)
    caption = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    return caption

# 画像URLを指定してキャプションを生成
image_url = "https://p-ch.jp/wp-content/uploads/2023/08/20230809_img_dfshduyrguyegfkh.jpg"  # ここに実際の画像URLを指定
caption = generate_caption(image_url)
print("Generated Caption:", caption)
```
